```python
# base import
import numpy as np
import matplotlib.pyplot as plt

# mock data
x = np.arange(0,100,1)
y = np.random.normal(0,10,100)

# create a new figure，1:0.618
plt.figure(figsize=(10,6.18))

# plot
# fmt = '[color][marker][line]'
# plt.plot(x,y,color='g',marker='.',markersize = '15',linestyle='--',linewidth='2')
#simple format
plt.plot(x,y,'g.--',markersize = '15',linewidth='2',label='line1')
plt.plot(x,x,'r.--',markersize = '15',linewidth='2',label='line2线')
# x,y axis settings
plt.xlabel('x-axis',fontproperties='SimHei',fontsize =25)
plt.ylabel('y-axis轴',fontproperties='SimHei',fontsize =25)
plt.xlim((0,100))
plt.ylim((-100,100))
plt.xticks(rotation=20,fontsize=15)
plt.yticks(fontsize=15)

# legend settings
plt.legend(loc='best',prop={'family':'SimHei','size':15})

# show
plt.grid()
plt.show()

```
